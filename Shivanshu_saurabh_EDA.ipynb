```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers_df = pd.read_csv("Customers.csv")
products_df = pd.read_csv("Products.csv")
transactions_df = pd.read_csv("Transactions.csv")
```

```python
print("\n--- Customers Dataset ---")
print("Basic Information:")
print(customers_df.info())
```

```python
print("\nSample Rows:")
print(customers_df.head())

print("\nMissing Values:")
print(customers_df.isnull().sum())

print("\nSummary Statistics:")
print(customers_df.describe())
```

```python
customer_numeric = customers_df.select_dtypes(include=['float64', 'int64'])
if not customer_numeric.empty:
    plt.figure(figsize=(10, 6))
    sns.heatmap(customer_numeric.corr(), annot=True, cmap="coolwarm")
    plt.title("Customers Dataset: Correlation Heatmap")
    plt.show()
else:
    print("No numeric columns available for correlation heatmap in Customers dataset.")
```

```python
print("\n--- Products Dataset ---")
print("Basic Information:")
print(products_df.info())
```

```python
print("\nSample Rows:")
print(products_df.head())
```

```python
print("\nMissing Values:")
print(products_df.isnull().sum())
```

```python
print("\nSummary Statistics (Numeric Columns):")
print(products_df.describe())
```

```python
print("\nUnique Product Categories:")
print(products_df["Category"].unique())
```

```python
print("\n--- Transactions Dataset ---")
print("Basic Information:")
print(transactions_df.info())
```

```python
print("\nSample Rows:")
print(transactions_df.head())
```

```python
print("\nMissing Values:")
print(transactions_df.isnull().sum())

print("\nSummary Statistics:")
print(transactions_df.describe())
```

```python
transaction_numeric = transactions_df.select_dtypes(include=['float64', 'int64'])
if not transaction_numeric.empty:
    plt.figure(figsize=(10, 6))
    sns.heatmap(transaction_numeric.corr(), annot=True, cmap="coolwarm")
    plt.title("Transactions Dataset: Correlation Heatmap")
    plt.show()
else:
    print("No numeric columns available for correlation heatmap in Transactions dataset.")
```

```python
merged_df = transactions_df.merge(customers_df, on="CustomerID", how="inner") \
                           .merge(products_df, on="ProductID", how="inner")
```

```python
print("\n--- Merged Dataset ---")
print("Basic Information:")
print(merged_df.info())
```

```python
print("\nSample Rows:")
print(merged_df.head())

print("\nMissing Values:")
print(merged_df.isnull().sum())
```

```python
if "Income" in merged_df.columns and "Amount" in merged_df.columns:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=merged_df, x="Income", y="Amount", hue="Category", palette="viridis")
    plt.title("Income vs Amount Spent by Product Category")
    plt.xlabel("Income")
    plt.ylabel("Amount Spent")
    plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=merged_df, y="Category", order=merged_df["Category"].value_counts().index, palette="pastel")
plt.title("Most Popular Product Categories")
plt.xlabel("Count")
plt.ylabel("Product Category")
plt.show()
```

```python
customers_df.describe().to_csv("Customers_EDA_Summary.csv")
products_df.describe().to_csv("Products_EDA_Summary.csv")
transactions_df.describe().to_csv("Transactions_EDA_Summary.csv")
merged_df.describe().to_csv("Merged_Dataset_EDA_Summary.csv")
```
